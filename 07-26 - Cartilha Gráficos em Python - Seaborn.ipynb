## Importação

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Importando Base de Dados

```python
df_videos = pd.read_excel('videosYT.xlsx','videos') #df de dataframe (comum utilizar dessa forma)
display(df_videos) 
```

## Tipos de Gráficos

Documentação : https://seaborn.pydata.org/api.html#distribution-api



## Criando um gráfico de dispersão:
Documentação : https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
# relplot --> grafico de dispersão e linha
# displot --> grafico de histograma
# categorical --> graficos de categorias

fig = sns.scatterplot(data=df_videos, x="Nº de Views", y="Nº de Likes",hue ='Responsável')
plt.show(fig)
```

## Criando um gráfico de dispersão(relacional):

Documentação : https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot


```python
fig = sns.relplot(data=df_videos, x="Nº de Views", y="Nº de Likes",hue = 'Categoria', col='Responsável')
plt.show(fig)
```

## Criando um gráfico de linha:
Documentação : https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```python
df_inscritos = pd.read_excel('videosYT.xlsx','Inscritos')

graf_linha = sns.lineplot(data=df_inscritos, x='Mês/Ano', y='Inscritos', color='blue')
plt.show(graf_linha)
```

## Criando Histogramas:

Documentação : https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot


```python
fig = sns.displot(data=df_videos, x="Nº de Views", col='Responsável')
# metodo hue vc compara duas coisas 
# metodo col divide duas coisas 

plt.show(fig)
```

## Criando uma regressão linear:

Documentação: https://seaborn.pydata.org/tutorial/regression.html#functions-to-draw-linear-regression-models

```python
graf_regre = sns.regplot(data = df_videos, x = "Nº de Views", y= "Nº de Likes" )
# regplot para 1 linha de visualização
plt.show(graf_regre)
```

```python
graf_regre = sns.lmplot(data = df_videos, x = "Nº de Views", y= "Nº de Likes", hue= 'Responsável')
#lmplot para mais de uma linha de visualização
plt.show(graf_regre)
```
